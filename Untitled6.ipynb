```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data={'course':['python','java','c','html'],
      'marks':[35,48,52,96]}
df=pd.DataFrame(data)
print(df)

# Creating the polar plot
courses = df['course']
marks = df['marks']
angles = np.linspace(0, 2*np.pi, len(courses), endpoint=False)
# close the plot
angles = np.concatenate((angles, [angles[0]]))
marks = np.concatenate((marks, [marks[0]]))

fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, marks, 'o-', linewidth=2)
ax.fill(angles, marks, alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi, courses)
plt.title('Polar Chart of Course Marks')
plt.show()
```
