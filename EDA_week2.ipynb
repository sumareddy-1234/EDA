```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={
    'year':[2016,2017,2019,2021,2022,2023],
    'sales':[250,260,270,300,310,450],
    'profit':[20,30,50,45,60,75],
    'Market_share':[22,10,12,16,18,7]
}
df1=pd.DataFrame(data)
df1.head()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df1['year'],df1['sales'],marker='o',color='b',label='sales')
plt.title('yearly sales Trends')
plt.xlabel('year')
plt.ylabel('sales')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df1['sales'],df1['profit'],marker='o',color='r')
plt.title('sales vs profit')
plt.xlabel('sales')
plt.ylabel('profit')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df1['sales'],df1['profit'],s=df1['Market_share']*10,alpha=0.5,color='purple')
plt.title('sales vs profit with market share')
plt.xlabel('sales')
plt.ylabel('profit')
plt.grid(True)
plt.show()
```

```python
data={
    'product':['mobiles','laptops','desktops','smartwatches'],
    'delhi':[500,410,390,50],
    'banglore':[500,300,360,50],
    'chennai':[500,340,320,50]
}
df=pd.DataFrame(data)
df.set_index('product',inplace=True)
print(df)
```

```python
#area plot
df.plot(kind='area',stacked=False,alpha=0.8,figsize=(8,5))
plt.title('Area Plot of product sales')
plt.xlabel('product')
plt.ylabel('sales')
plt.legend(title="Regions")
plt.show()
```

```python
#Stackedplot
df.plot(kind='area',stacked=True,alpha=0.8,figsize=(8,5))
plt.title('Area Plot of product sales')
plt.xlabel('product')
plt.ylabel('sales')
plt.legend(title="Regions")
plt.show()
```

```python
#Pie Chart
total_sales=df.sum();
total_sales.plot(kind='pie',autopct='%1.1f%%',startangle=90,figsize=(6,6),explode=[0.050,0,0])
plt.title('Sales Distribution across regions')
plt.ylabel('')
plt.show()
```

```python
#Table Chart
from pandas.plotting import table
fig,ax=plt.subplots(figsize=(6,3))
ax.set_axis_off()
table(ax,df,loc='center',cellLoc='center',colWidths=[0.2]*len(df.columns))
plt.title('Sales Data Table')
plt.show()
```
