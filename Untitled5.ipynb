```python
#create a sample
import pandas as pd
data={"name":["dolly","julie","sofia","elsa"],
      "course":["java","python","c","c++"],
      "marks":[95,100,85,90]}
df=pd.DataFrame(data)
df
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(2,8))
sns.histplot(data,bins=20,kde=True)
plt.xlabel("course")
plt.ylabel("marks")
plt.title("Histogram")
plt.show()
```
