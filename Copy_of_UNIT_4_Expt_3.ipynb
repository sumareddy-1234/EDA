```python
#Expt 3: Compute measure of Central Tendency on a sample dataset
# a)Mean b)Median c)Mode
#Explore Measures of Dispersion on a sample dataset
#a)variance  b)standard deviation c)skewness d)kurtosis.
```

```python
import pandas as pd
df=pd.read_csv("/content/Bengaluru_House_Data (2) (4).csv")
df.head()
```

```python
def compute_statistics(column):
  mean_value=df['price'].mean()
  median_value=df['price'].median()
  mode_value=df['price'].mode()
  variance_value=df['price'].var()
  std_value=df['price'].std()
  skewness_value=df['price'].skew()
  kurtosis_value=df['price'].kurtosis()
  print(f"Statistics for {column}:")
  print("Mean:",mean_value)
  print("Median:",median_value)
  print("Mode:",mode_value)
  print("Variance:",variance_value)
  print("Standard Deviation:",std_value)
  print("Skewness:",skewness_value)
  print("Kurtosis:",kurtosis_value)
#Apply the function to each numerical column
numerical_columns=['price','total_sqft','bath','balcony']
for col in numerical_columns:
  compute_statistics(col)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Generate 1000 random values from a Uniform Distribution between 10 and 50
uniform_data=np.random.uniform(low=10,high=50,size=1000)
#plot the histogram
sns.histplot(uniform_data,label="Uniform Distribution",kde=True,bins=30,color="blue")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Uniform Distribution")
plt.legend()
plt.show()
```

```python
#normal distribution(bell shaped curve)
import numpy as np
import matplotlib.pyplot as plt
mean=100
standard_deviation=5
size=100000
#creating a normal distribution data
values=np.random.normal(mean,standard_deviation,size)
#plotting the histogram
plt.hist(values,100)
#plotting mean line
plt.axvline(values.mean(),color='k',linestyle='dashed',linewidth=2)
plt.show()
```
