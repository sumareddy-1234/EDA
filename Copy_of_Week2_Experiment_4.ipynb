```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data={'courses':['python','java','c','c++','HTML'],
      'values':[10,100,95,85,100]}
df=pd.DataFrame(data)
df
```

```python
#polar chart setup
categories=df['courses']
values=df['values'].tolist()
angles=np.linspace(0,2*np.pi,len(categories),endpoint=False).tolist()
values+=values[:1]
angles+=angles[:1]
fig,ax=plt.subplots(figsize=(6,6),subplot_kw={'projection':'polar'})
ax.fill(angles,values,color='skyblue',alpha=0.4)
ax.plot(angles,values,marker='o',color='blue')
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title('polar chart')
plt.show()
```

```python
data={
    'sno':[501,502,503,504,505,506,507,508,509,510],
    'marks':[45,67,34,89,56,42,99,67,87,46]
}
df=pd.DataFrame(data)
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
random_marks=np.random.normal(loc=np.mean(df['marks']),
                              scale=np.std(df['marks']),
                              size=100)
#plot the histogram
plt.hist(random_marks,bins=3,color='skyblue',edgecolor='black',alpha=0.7)
plt.title('Histogram of Random Marks Data')
plt.xlabel('Marks Range')
plt.ylabel('Frequency')
plt.grid(axis='y',linestyle='--',alpha=0.7)
#show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
#sample data
categories=['Product A','Product B','Product C','Product D','Product E']
values=[50,80,30,60,90]
#create the lollipop chart
plt.stem(categories,values,linefmt='gray',markerfmt='o',basefmt='')
#add labels and title
plt.xlabel("products")
plt.ylabel("Sales (in $1000)")
plt.title("Lollipop chart of Product sales")
plt.show()
```
